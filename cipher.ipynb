```python
# Import necessary libraries
import rasterio
from rasterio.enums import Resampling
import matplotlib.pyplot as plt
import numpy as np

# Path to your GeoTIFF file
path = '/home/user/.cache/huggingface/hub/datasets--pints-sig--Land_Cover_Map_2023_10m_classified_pixels_GB/snapshots/93682d97808ea967159f6372d1805723fdc92b29/FME_61343034_1740838768029_7621/data/7727ce7d-531e-4d77-b756-5cc59ff016bd/gblcm2023_10m.tif'

# Open the file with rasterio
with rasterio.open(path) as src:
    # Calculate a scale factor to make the image manageable
    # Let's aim for a final image of around 1000-2000 pixels in the largest dimension
    scale_factor = max(src.width, src.height) / 1500
    
    # Calculate new dimensions
    new_width = int(src.width // scale_factor)
    new_height = int(src.height // scale_factor)
    
    print(f"Original size: {src.width} x {src.height}")
    print(f"Resampled size: {new_width} x {new_height}")
    print(f"Scale factor: {scale_factor:.1f}x")
    
    # Read the data with resampling
    data = src.read(
        out_shape=(src.count, new_height, new_width),
        resampling=Resampling.nearest
    )
    
    # Update transform for the new resolution
    transform = src.transform * src.transform.scale(
        (src.width / new_width),
        (src.height / new_height)
    )
    
    # Display band 1 (assuming it's the classification data)
    plt.figure(figsize=(12, 8))
    plt.imshow(data[0], cmap='viridis')
    plt.colorbar(label='Land Cover Class')
    plt.title('Land Cover Map 2023 (Subsampled Overview)')
    plt.axis('off')
    plt.show()
    
    # If you want to display band 2 as well
    plt.figure(figsize=(12, 8))
    plt.imshow(data[1], cmap='plasma')
    plt.colorbar(label='Band 2 Value')
    plt.title('Land Cover Map 2023 - Band 2 (Subsampled Overview)')
    plt.axis('off')
    plt.show()
    
    # Create an RGB composite using the two bands
    # This might look strange depending on what the bands represent
    rgb = np.zeros((3, new_height, new_width), dtype=data.dtype)
    
    # Normalize data for visualization
    band1_norm = (data[0] - data[0].min()) / (data[0].max() - data[0].min() + 1e-10)
    band2_norm = (data[1] - data[1].min()) / (data[1].max() - data[1].min() + 1e-10)
    
    rgb[0, :, :] = band1_norm  # Red channel
    rgb[1, :, :] = band2_norm  # Green channel
    rgb[2, :, :] = (band1_norm + band2_norm) / 2  # Blue channel as average
    
    # Transpose for matplotlib's expected shape
    rgb = np.transpose(rgb, (1, 2, 0))
    
    # Clip values to valid range
    rgb = np.clip(rgb, 0, 1)
    
    # Plot RGB composite
    plt.figure(figsize=(12, 8))
    plt.imshow(rgb)
    plt.title('Land Cover Map 2023 (RGB Composite)')
    plt.axis('off')
    plt.show()
```
